# Замена пропущенных значений

```python
string_columns = [
    "Critic Consensus",
    "Directors",
    "Authors",
    "Actors",
    "Production Company",
    "Audience Status"
]

value_columns = [
    "Tomatometer Rating",
    "Tomatometer Count",
    "Audience Rating",
    "Audience Count",
    "Tomatometer Top Critics Count",
    "Tomatometer Fresh Critics Count",
    "Tomatometer Rotten Critics Count"
]

for row in string_columns:
    data[row] = data[row].fillna('unknown')

for row in value_columns:
    data[row] = data[row].fillna(data[row].median())
```

```python
del data['Unnamed: 0']
data.dropna(subset=['Original Release Date', 'Streaming Release Date'], inplace=True)
```

# Изучение и обработка выбросов

```python
outlier = data[['Movie Time','IMDB Rating','Tomatometer Rating','Tomatometer Count','Audience Rating']]
outlier
```

```python
# Определить нижние пределы для выбросов
Q1 = outlier.quantile(0.25)
Q1
```

```python
# Определить верхние пределы выбросов
Q3 = outlier.quantile(0.75)
Q3
```

```python
# IQR-оценки
IQR = Q3-Q1
# Отфильтрованные данные, из которых удалены выбросы
data_filtered = outlier[~((outlier < (Q1 - 1.5 * IQR)) |(outlier > (Q3 + 1.5 * IQR))).any(axis=1)]
data_filtered
```

```python
# Копирование индексов
index_list = list(data_filtered.index.values)
```

```python
# Хранить данные без выбросов в отдельном DataFrame.
data_filtered = data[data.index.isin(index_list)]
data_filtered.head()
```

```python
# создать гистограмму с отфильтрованным выбросом
filtered_outlier = data_filtered[['Movie Time','IMDB Rating','Tomatometer Rating','Tomatometer Count','Audience Rating']]

for col in filtered_outlier:
    data_filtered.hist(col, bins=100)
```

# Вывод:
1. Определили нижний и верхний пределы выбросов, определив 25-й и 75-й процентили. Затем нашли межквартильный размах. Использовали эту информацию для фильтрации выбросов. После удаления выбросов график выглядит более симметричным по сравнению с выбросами, и в данных легче наблюдать среднее значение, медиану и моду.


2. Судя по гистограмме времени, большая часть фильмов распределена в диапазоне 100-140. Максимальное время — 180.


3. Судя по гистограмме значений столбца 'IMDB Rating', самая популярная оценка находится в диапозоне от 7.0-8.0
4. Судя по гистограмме значений столбца 'Tomatometer Rating', самая популярная оценка-93.
5. Судя по гистограмме значений столбца 'Tomatometer Count', самое популярное кол-во находится в диапозоне 0-100.
6. Судя по гистограмме значений столбца 'Audience Rating', самая популярная оценка-87.
