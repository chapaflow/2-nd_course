```python
import math
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from scipy import linalg, stats
```

# Моя функция

```python
def f(x):
  return math.tan(x)/math.sin(x)
```

# task№1

```python
a = 0.1
b = 1
step = b/10

ans = []
pts = []

while a < b:
  pts.append(a)
  ans.append(f(a))
  a += step

for i in range (0, len(ans)):
  print(f'При значении {round(pts[i], 1)}: {round(ans[i], 2)}')
```

# task№2

```python
plt.title('Моя функция',fontsize=20)
plt.plot(pts,ans)
SMALL_SIZE = 8
MEDIUM_SIZE = 10

plt.rc('font', size=SMALL_SIZE)
plt.rc('axes', titlesize=SMALL_SIZE)
plt.rc('axes', labelsize=MEDIUM_SIZE)
plt.rc('legend', fontsize=MEDIUM_SIZE)


plt.legend(['tan(x)/sin(x)'])
plt.ylabel('y (ans)')
plt.xlabel('x (pts)')

plt.show()
```

# task№3

```python
plt.plot(pts, ans,marker='.', markerfacecolor=[1, 0, 0, 1], markersize=11)
plt.title('Моя функиця', fontsize=20)
plt.grid(color=([0.3,0.2,1]), linestyle='-', linewidth=0.3, alpha=1)

plt.ylabel('y (ans)')
plt.xlabel('x (pts)')

plt.show()
```

# task№6

```python
def func(x, y):
    return (x - 4) + (y - 3)**2

x = np.random.random(5)
y = np.random.random(5)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot(x,y, func(x,y), color=[1, 0, 0, 1])

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x)')
ax.set_title('Трехмерный график', fontsize = 20)

plt.show()
```
